```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset (assuming you have it as a CSV file)
# Replace 'titanic.csv' with the actual path to your dataset file
titanic_data = pd.read_csv('tested.csv')
```

```python
# Choose two continuous variables from the dataset
# For example, let's use 'Age' and 'Fare'
continuous_variable1 = 'Age'
continuous_variable2 = 'Fare'
```

```python
# Extract the selected variables
x_values = titanic_data[continuous_variable1]
y_values = titanic_data[continuous_variable2]
```

```python

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, color='b')
plt.title(f"Relationship between {continuous_variable1} and {continuous_variable2}")
plt.xlabel(continuous_variable1)
plt.ylabel(continuous_variable2)
plt.grid(True)
```

```python
# Show the plot
plt.show()
```
