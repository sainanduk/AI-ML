```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('tested.csv')
```

```python
# Display basic information about the dataset
print("Basic Info about the Titanic Dataset:")
print(titanic_df.info())
```

```python
# Check for missing values
print("\nMissing Values:")
missing_values = titanic_df.isnull().sum()
print(missing_values)
```

```python
# Summary statistics for numerical features
print("\nSummary Statistics for Numerical Features:")
print(titanic_df.describe())
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
correlation_matrix = titanic_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python


# Variance of numerical features
print("\nVariance of Numerical Features:")
variance = titanic_df.var()
print(variance)
```

```python
# Box plots for numerical features to identify outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic_df.drop(columns=['PassengerId', 'Pclass', 'Survived', 'SibSp', 'Parch']),
            palette='Set2', orient='h')
plt.title("Box Plots of Numerical Features")
plt.show()
```

```python
# Count plots for categorical features
categorical_features = ['Pclass', 'Sex', 'Embarked']
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=titanic_df, x=feature, hue='Survived', palette='Set2')
    plt.title(f'Count Plot of {feature}')
    plt.show()
```

```python


# Pairplot for selected numerical features
numerical_features = ['Age', 'Fare']
sns.pairplot(titanic_df, hue='Survived', vars=numerical_features, palette='Set2')
plt.title("Pairplot of Numerical Features")
plt.show()
```
