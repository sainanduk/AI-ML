```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the Iris dataset from Seaborn
iris = pd.read_csv("Iris.csv")
```

```python

# Create a correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
```

```python
# Generate two more visuals of your choice
# Visual 1: Pairplot
plt.figure(figsize=(10, 8))
sns.pairplot(iris, hue="Species", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset")
```

```python

# Visual 2: Bar chart of species counts
species_counts = iris["Species"].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=species_counts.index, y=species_counts.values)
plt.title("Species Counts")
plt.xlabel("Species")
plt.ylabel("Count")

# Show the plots
plt.show()
```
